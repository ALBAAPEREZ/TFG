```python
import os

train_dir = r'C:\TFG\Dataset_Final_Procesado\Dataset_Final_Procesado\train'
test_dir  = r'C:\TFG\Dataset_Final_Procesado\Dataset_Final_Procesado\test'

def extraer_id_video(nombre_archivo):
    # Elimina la extensión y separa por "_f" para quedarse con el ID del vídeo
    sin_ext = os.path.splitext(nombre_archivo)[0]  # quita .jpg
    # El ID es todo menos la última parte "_fX"
    partes = sin_ext.rsplit('_f', 1)  # divide por el ÚLTIMO "_f"
    return partes[0]  # ej: "DeepFakeDetection_22_23__outside_talking_pan_laughing__TU7RHG0S"

# --- Recoger IDs de train ---
ids_train = set()
for root, dirs, files in os.walk(train_dir):
    for f in files:
        if f.lower().endswith(('.jpg', '.png')):
            ids_train.add(extraer_id_video(f))

# --- Recoger IDs de test ---
ids_test = set()
for root, dirs, files in os.walk(test_dir):
    for f in files:
        if f.lower().endswith(('.jpg', '.png')):
            ids_test.add(extraer_id_video(f))

# --- Comprobar solapamiento ---
solapamiento = ids_train & ids_test

print(f"Vídeos únicos en train: {len(ids_train)}")
print(f"Vídeos únicos en test:  {len(ids_test)}")
print(f"Vídeos en AMBOS splits: {len(solapamiento)}")

if len(solapamiento) == 0:
    print("\n✅ Split correcto: ningún vídeo está en train y test a la vez.")
    print("   Tu entrenamiento NO tiene data leakage.")
else:
    print(f"\n❌ HAY DATA LEAKAGE en {len(solapamiento)} vídeos.")
    print("Ejemplos de vídeos repetidos:")
    for v in list(solapamiento)[:5]:
        print(f"  → {v}")
```
